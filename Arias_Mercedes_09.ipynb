1. Genere una lista de números desde 0 a 50 (incluido) con una distancia de 2 entre cada numero, tome cada uno de estos numeros como `x` y genere una gráfica de los datos para las funciones:
    - $y = x^2$
    - $y = x^3 - x^2$
    
    - puede usar `x = np.arange(0,51, 2).astype(int)`

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 51, 2).astype(int)
y1 = x**2
y2 = x**3 - x**2
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label = "y1 = x²", color = "purple", marker = "o")
plt.plot(x, y2, label = "y2 = x³-x²", color = "green", marker = "s")

plt.title("Gráfica de y1 = x² y y2 = x³-x²", fontsize=15)
plt.xlabel("x", fontsize=13)
plt.ylabel("y", fontsize=13)
plt.legend()
plt.grid(True, linestyle = "-", alpha = 0.5)

plt.show()
````

2. Dado que se puede generar una distribución normal estándar con;
``` python
mu = 100
sigma = 15
data = mu + (sigma * np.random.randn(10000))
```
Dibuje un histograma de 100 contenedores, donde las barras son verticales, elija dos barras y modifique su color por uno diferente al resto del histograma

````python
import numpy as np
import matplotlib.pyplot as plt

mu = 100  
sigma = 15  
data = mu + (sigma * np.random.randn(10000)) 


fig, ax = plt.subplots(figsize=(10, 6))
counts, bins, patches = ax.hist(data, bins=100, color="pink", edgecolor="black")


highlight_indices = [25, 50]  
for idx in highlight_indices:
    patches[idx].set_facecolor("red")  
    patches[idx].set_edgecolor("blue")  


ax.set_title("Histograma de distribución normal estándar", fontsize=14)
ax.set_xlabel("Valor", fontsize=12)
ax.set_ylabel("Frecuencia", fontsize=12)

plt.show()
````

3.  Grafique la función $tan(x)$ entre los valores de -$\pi$ y $\pi$ con los límites de $y$ de -10 y 10

````python
import numpy as np
import matplotlib.pyplot as plt

desde = -np.pi
hasta = np.pi
x = np.linspace(desde, hasta, 1000)
y = np.tan(x)


plt.figure(figsize=(8, 6))
plt.plot(x, y, color="orange", label="tan(x)", linestyle="-")


plt.ylim(-10, 10)  
plt.axhline(0, color="black", linewidth=0.8, linestyle="-")  
plt.axvline(0, color="black", linewidth=0.8, linestyle="-")  
plt.title("Gráfica de tan(x) entre -pi y pi", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
````

4. Genere una distribución normal con centro en 10, desviación estandar en 5 y tamaño 10000. Luego grafique tres histogramas de dicha distribución con 100, 50 y 10 bins, orientación vertical. Los tres histogramas deben estar en una sola fila en tres columnas, tenga cuidado de que las etiquetas de los ejes no esten superpuestas. Nombre el eje y, coloque un tamaño aprox de 9x2 pulgadas y 200dpi. Puede agregar/editar otras características de los plots segun lo crea conveninente.

````python
mu = 10  
sigma = 5  
data = np.random.normal(mu, sigma, 10000)

fig, axes = plt.subplots(1, 3, figsize=(9, 2), dpi=200)
bins_list = [100, 50, 10]

for ax, bins in zip(axes, bins_list):
    ax.hist(data, bins=bins, color="pink", edgecolor="black")
    ax.set_title(f"{bins} Bins", fontsize=10)
    ax.set_xlabel("Valor", fontsize=8)
    ax.set_ylabel("Frecuencia", fontsize=8)
    ax.tick_params(axis='both', which='major', labelsize=8)


tight_layout = fig.tight_layout()
plt.show()
````
